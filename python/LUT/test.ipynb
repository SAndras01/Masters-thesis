```python
def pow_3_5_7_9(num):
    pow_3 = num*num*num
    pow_5 = pow_3*num*num
    pow_7 = pow_5*num*num
    pow_9 = pow_7*num*num
    return pow_3,pow_5,pow_7,pow_9
```

```python
def sin_taylor(x, scaler_param):
    scaler_pows = pow_3_5_7_9(scaler_param)
    x_pows = pow_3_5_7_9(x)
    return  x/scaler_param \
            - x_pows[0]/(6*scaler_pows[0]) \
            + x_pows[1]/(120*scaler_pows[1]) \
            - x_pows[2]/(5040*scaler_pows[2]) \
            + x_pows[3]/(362880*scaler_pows[3])
```
